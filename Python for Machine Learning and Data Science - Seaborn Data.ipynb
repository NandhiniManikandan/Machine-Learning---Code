### Distribution plot
* Displays a single continous figure and help visualize properties such as deviation and average values.
* There are 3 main distribution plot type:
    * Rug Plot 
        * Rug plot is the simplest distribution plot and merely adds a dash or tick line for every single value
        * The y-axis does not have any meaning.
    * Histogram
        * If we count how many ticks are present per various x-ranges, we can create a histogram.
        * We have to choose no of bins. Bin is the no of segments with which the x-axis is cut.
        * we count the number of ticks within the bins, based on which we get the histogram.
    * Kernel Density Estimation (KDE) Plot
        * It is a method of estimating a probability density function of a random variable
        * In simpler terms, it is a way of estimating a continous probability curve for a finite data sample

### CATEGORICAL PLOT - STATISTICS WITHIN CATEGORY
    * The categorical plots discussed here will display a statistical metrics per a category.
    * For example: mean value per a category or a count of the number of rows per category.
    * It is a visualization equivalent of a groupby() call
 * The two main types of plots for this are:
    * countplot() : counts number of rows per category
    * barplot() : General form of displaying any chosen matrics per category
* The barplot is the general form that allows you to choose any measure or estimators for the y-axis
* We could plot the mean value and standard deviation per category instead.

** Important Note!
* Be very careful with these plots, since the bar is filled and continous, a viewer may interpret continuity along the y-axis which may be incorrect!
* Always make sure to add additional labeling and explation for these plots.

### CATEGORICAL PLOTS - DISTRIBUTION WITHIN CATEGORIES

* We will first seperate out each category, then create the distribution visualization.
* Distribution within Categories:
    * Boxplot
    * Violin Plot
    * Swarmplot
    * Boxenplot (Letter-Value plot)
##### BOX PLOT
* The Boxplot displays the distribution of a continous variable
* It does it through the use of quartiles.
* Quartiles seperate out the data into 4 equal number of data points.
    * 25% of data points are in bottom quartile.
    * 50th percentile (Q2) is the median.
##### VIOLIN PLOT
* The violin plot plays a similar role as the box plot
* It displays the probability density across the data using a KDE plot
* We can imagine it as the mirrored KDE Plot
##### SWARM PLOT
* The swarm plot is very simple and it just shows all the data points in the distribution.
* For a very large data set it won't show all the points, but will display the general distribution of them.
##### BOXEN PLOT
* The boxen plot is designed as an expansion upon the normal box plot
* Use the plot only if you and your audience is sure about this coz when we look it up in google it autocorrects it as boxplot.

#### COMPARISON PLOTS
* Comparison plots are essentially 2D versions of the plots .
* The two main plot types discussed here:
    * jointplot()
    * Pairplot()

#### JOINTPLOT()
    * We can map histogram to  each feature of a scatterplot to clarify the distribution withtin each feature.
    * We can also adjust the scatterplot to be a hex plot or a 2D KDE Plot
    
Note: Hexagons in the hex plot are dark when there are more points that fall into that area.
2D KDE shows a shaded distributions between both KDE's.

#### PAIR PLOT()
    * The Pairplot() is a quick way to compare all numerical columns in a Data Frame
    * It automatically creates a histogram for each column and scatterplot comparison between all possible combinations of columns
    
* Warning:
         * Pairplot() can be CPU and RAM intensive for large dataframes with many columns
         * It is a good idea to first filter down only the columns you are interested in.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("StudentsPerformance.csv")
df.head()
```

```python
# Note we need continous range of two data sets for the jointplot()
sns.jointplot(data=df,x='math score', y='reading score') # The histogram on top is the math score variation and the graph on the right is the reading score variation and the points in the middle are where both the scores intersect
```

```python
sns.jointplot(data=df,x='math score', y='reading score', kind= 'hex') # Kind gives the way in which we want to see the plot, and hexagonal plot 
```

```python
sns.jointplot(data=df,x='math score', y='reading score', kind = 'kde', shade = True) # Shows the plot as a 2D KDE with the shade color. Note we can always change the shade using the palatte keyword
```

```python
sns.jointplot(data=df,x='math score', y='reading score', hue = 'gender') # It Creates two different KDE's per the hue specified per continous feature along with the points
```

```python
sns.pairplot(data=df)# Creates a pair grid and along the diagonal we have the histogram for each column
```

```python
sns.pairplot(data=df, hue = 'gender') # When we add in the hue, the diagonal is changed in to a KDE automatically as the stacked histogram is difficult to read
```

```python
sns.pairplot(data=df, hue = 'gender', diag_kind= 'hist') # This diag_kind= 'hist' shows diagonals as histograms itselfs
```

```python
sns.pairplot(data=df, hue = 'gender', corner = True) # Gets rids of the duplicates and only shows the unique datas
```

### GRID PLOT
* Seaborn grid calls uses Matplotlib subplots() to automatically create a grid based off a categorical column
* Instead ao passing in a specific number of columns or rows for the subplots, we can simply pass in the name of the column and seaborn will automatically map the subplots grid
* Many of seaborns built-in-plot calls are running on top of the grid system
* Directly calling the grid system allows the users to heavily customize the plots.

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("StudentsPerformance.csv")
df.head()
```

```python
sns.catplot(data=df, x ='gender',y ='math score', kind = 'box') # This looks a normal box plot but with the extra categories applied it will be more clear
```

```python
sns.catplot(data=df, x ='gender',y ='math score', kind = 'box', row = 'lunch') # Takes the box plot and creates grids for every category of lunch. This is similar to hue but it here it creates a row wise grid lined one below the other
```

```python
sns.catplot(data=df, x ='gender',y ='math score', kind = 'box', hue= 'lunch') # This is the difference between hue an drow. In hue we see both the lunch value for the female category together but in row the hue is mentioned out for both the categories seperately
# Note: More the catrgories more difficult it is going to be for us to read
```

```python
sns.catplot(data=df, x ='gender',y ='math score', kind = 'box', col= 'lunch' , row ='test preparation course') # Here the labelling is clear.
```

```python
sns.pairplot(data=df) # Creates the Pairplot based on the pair grid functionality
```

```python
g = sns.PairGrid(df) # Creates a grid and stores it in g as a template

g= g.map_upper(sns.scatterplot) # Maps the upper triangle with the scatter plot but make sure that the x and y values are continous for the scatterplot to work.

g = g.map_lower(sns.kdeplot) #Maps the KDE Plot to the lower triangle

g= g.map_diag(sns.histplot) # Make sure that the diagonal is a single value plot as the (math score vs math score is of no sense when we use two continous values)
```

```python
g = sns.PairGrid(df, hue= 'gender') # Creates a grid and stores it in g as a template and has the hue as gender and this is applied to all the subplots

g= g.map_upper(sns.scatterplot) # Maps the upper triangle with the scatter plot but make sure that the x and y values are continous for the scatterplot to work.

g = g.map_lower(sns.kdeplot) #Maps the KDE Plot to the lower triangle

g= g.map_diag(sns.histplot) # Make sure that the diagonal is a single value plot as the (math score vs math score is of no sense when we use two continous values)

g= g.add_legend() #this doesn't have legend so we have to add it manually
```

### MATRIX PLOTS
* Matrix Plots are the visual equivalents of displaying a pivot table
* The matrix plot displays all the data passed in, visaulizing all the numeric values in a Data Frame
* Note:
    * Not every Data Frame is valid choice for a matrix plot such as heat map
* The two main types of matrix plots are
    * heatplot()
        * Visually displaying the distribution of cell values with color mapping.
    * clustermap()
        * Same visual as a heat map but first conducts hierarchical clustering to reorganize data into groups.
##### NOTE
* Note that heatmap should ideally have all cells be in the same units, so the color mapping makes sense across the entire DF

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```
