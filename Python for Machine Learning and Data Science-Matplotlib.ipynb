### MATPLOTLIB- UNDERSTANDING THE FIGURE OBJECT

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=np.linspace(0,10,11) #Getting 11 datas linearlyspaced by 0 and 10
b=a**4 #Getting the 4thpower of each value in a
print(a)
print(b)
```

```python
x=np.arange(0,10)
y=2*x
print(x)
print(y)
```

```python
fig=plt.figure() #Creating a empty plot
axes=fig.add_axes([0,0,1,1]) #adding the axes to the plot
axes.plot(x,y) #Sending the variables that we want to plot
```

```python
fig=plt.figure() #Creating a empty plot
axes=fig.add_axes([0,0,1,1]) #adding the axes to the plot
axes.plot(a,b) #Plots the variables a and b on the plot
```

```python
type(fig)
```

```python
fig=plt.figure()

#LARGE AXIS
axes1= fig.add_axes([0,0,1,1])
axes1.plot(a,b)

axes1.set_xlim(0,8)  #Note that we would have to use set_ when we are using it on a set of axis.and this labelling is added only to the larger plot and doesn't affect the inner set of axes
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of four')


#SMALL AXES
axes2= fig.add_axes([0.2,0.2,0.5,0.5]) #Moves the plot 20% in on X and 20% up on Y. The o.5 represents that the width and height are half of the other axis  
axes2.plot(x,y)
axes2.set_xlim(1,2)  #Note that we would have to use set_ when we are using it on a set of axis.and this labelling is added only to the larger plot and doesn't affect the inner set of axes
axes2.set_ylim(0,10)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed in')
plt.show()
```

### MATPLOTLIB-SUBPLOT FUNCTIONALITY

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=np.linspace(0,10,11)
b=a**4
x=np.arange(0,10)
y=x*2
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2) #This has one rows and 2 columns
```

```python
type(axes) #this has a shape now as it is a numpy ndarry
```

```python
axes
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y) #Plots the x and y variables in the first box
axes[1].plot(a,b) #Plots the a and b variables in the second box
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)

axes[0].plot(x,y) 
axes[1].plot(a,b)  

#This shows an array as the shape has been changed
```

```python
axes.shape #shows that it is a 2D array
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)

axes[0][0].plot(x,y) #Axes[row][column].plot(varibales that we want to plot)
axes[1][0].plot(a,b)  
```

```python
fig,axes=plt.subplots(nrows=3,ncols=1) #For the 1D array where either the column or row is 1, we can loop through the axes and plot
for ax in axes:
    ax.plot(x,y) #Plots the same on all three plots
```

```python
fig,axes=plt.subplots(nrows=3,ncols=1) 
axes[0].plot(x,y)
plt.tight_layout() #Automatically adjust subplot parameters to give specified padding. That is it makes sure that the plot isn't repeated on the other canvas
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[0][1].plot(x,y)
axes[1][0].plot(x,y)
axes[1][1].plot(a,b)

#plots the variables on all the canvas and we can see that the plot line overlaps with the axis labels, using plt.tight_layout(), fixes to only some degree.
fig.subplots_adjust(wspace=1,hspace=0.5)#Hspace and the Wspace are the height and the width space respectively reserved for space between subplots and the way it is expressed is as a fractionpf the averages axis height and width repectively.
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[0][0].set_xlabel('X LABEL 0.0')
axes[0][0].set_ylabel('Y LABEL 0.0')
axes[0][0].set_title('TITLE 0.0')
axes[0][1].plot(x,y)
axes[0][1].set_xlabel('X LABEL 0.1')
axes[0][1].set_ylabel('Y LABEL 0.1')
axes[0][1].set_title('TITLE 0.1')
axes[1][0].plot(x,y)
axes[1][0].set_xlabel('X LABEL 1.0')
axes[1][0].set_ylabel('Y LABEL 1.0')
axes[1][0].set_title('TITLE 1.0')
axes[1][1].plot(a,b)
axes[1][1].set_xlabel('X LABEL 1.1')
axes[1][1].set_ylabel('Y LABEL 1.1')
axes[1][1].set_title('TITLE 1.1')
plt.tight_layout()
fig.suptitle("FIGURE LEVEL",fontsize=16) #Adds a figure level title of the size 16. Note the Figure level title is for the overall plot
#We can loop through these and we can also do all the labelling and limitting in all the array as well.
fig.savefig('New_Support.png',bbox_inches='tight')
```

### MATPLOTLIB - STYLING - LEGENDS

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig=plt.figure() #Create a new figure
x=np.linspace(0,10,11) #Creating a variable which has 11 values linearly spaced bwteen 0 and 10
ax=fig.add_axes([0,0,1,1]) #Adding the axes to the plot
ax.plot(x,x,label='X vs X') #Plots x by x on the canvas and adds the labelname to the plot
ax.plot(x,x**2,label='X vx X^2') # plots x by the square of x on the same canvas and adds the labelname to the plot
ax.legend(loc=(1.1,0.5)) #adds legend with the label name and the plot color to the canvas and the loc gives the location to which the legend should be placed so that it doesn't overlap with th plot
#Note that we can either use "Upper left" or give the location manually by giving the location value, and loc = 0 would let matplotlib choose the location of the legend
```

### MATPLOTLIB - STYLING - COLORS AND STYLES

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.linspace(0,10,11) #Creates a variable and has 11 values that are linearly spaced
fig=plt.figure() #Creates a new figure
ax=fig.add_axes([0,0,1,1]) #Adds axis to the figure
ax.plot(x,x,color='purple',label='X vs X') #plots "X vs X" and has the plot line as purple or any basic color that is given in quotes
#We can add our own color using the RGB HEX code that we can get from google
ax.plot(x,x+1,color='#0524ab',label='X vs X+1',lw=6) #Plots "X vs X+1" and has the plot line in our customized color and the lw is used to specify the linewidth,(we can also use lw=3 or linewidth=3)
ax.legend(loc=(0)) #Adds the legend to the plot
```

```python
x= np.linspace(0,10,11)
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x,color='purple',label='X vs X',ls='--') #The ls is used to specify the linestyle. We can either us ls or linestyle
# Possible linestyle options '--','-','-.',':','steps'
ax.plot(x,x+1,color='#0524ab',label='X vs X+1',lw=6)
ax.legend(loc=(0))
```

```python
x= np.linspace(0,10,11)
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x,color='purple',label='X vs X',ls='--')
lines=ax.plot(x,x+1,color='#0524ab',label='X vs X+1',lw=6)
ax.legend(loc=(0))
```

```python
x= np.linspace(0,10,11)
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x,color='purple',label='X vs X',ls='--')
lines=ax.plot(x,x+1,color='#0524ab',label='X vs X+1',lw=2)
lines[0].set_dashes([5,2,10,2]) #this is to create a custom linestyle where it has 5 solid points and 2 blank points and then 10 solid points and 2 blank points
```

```python
x= np.linspace(0,10,11)
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x,color='purple',label='X vs X',ls='--') 
ax.plot(x,x+1,color='#0524ab',label='X vs X+1',lw=6,marker='o',ls=':',ms=15,markerfacecolor='red',markeredgewidth=4,markeredgecolor='orange') #Adds the markers on the points, the marker parameter can be obtained from google and we can use ms or marker size to specify the size of the marker required
#Note that we can mentiont the marker edge size and the marker edge color aswell
ax.legend(loc=(0))
```
