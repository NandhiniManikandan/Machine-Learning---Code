### Conditional Filtering
* Conditional Filtering allows us to select rows based a condition on a column. 
* Typically in data analysis our datasets are large enough that we don't filter based on position, but instead based on condition.
* Columns are features or an attribute that is some sort of property or characteristics of every single instance
* Rows are instances of data

### Conditional Filtering can be used to:
* Filters by single Condition
* Filter by Multiple Condition
* Cneck against multiple possible values

```python
import numpy as np
import pandas as pd
```

```python
df= pd.read_csv("tips.csv") #Reading the file tips.csv
```

```python
df.head()
```

```python
df['total_bill']>40 #Gives if the Total_bill for each entry is greater than 40 or not
```

```python
bool_series=df['total_bill']>40 #Assigning the conditional statement to the variable
df[bool_series] #This displays the dataframe(s) for which the condition is true; 
```

```python
df[df['total_bill']>40] #Gives the same result as above
```

```python
df[df['sex']=="Male"] #Displays the datas where the sex is male
```

```python
df[df['size']>3] #Displays the datas for which the size is greater than 3
```

##### FOR MULTIPLE CONDITIONING
* AND & -- Both the conditions are to be true
* OR | -- Either of the two conditions are to be true
* Note the condition is to be within paranthesis

```python
df[(df['total_bill']>30)&(df['sex']=='Male')] #Displays the entries in which the total bill is greater than 30 and the sex is male. The 'and' isn't used as it shows error coz it work on only one boolean value
```

```python
df[(df['total_bill']>30)|(df['sex']=='Male')] #Displays the entries for which the sex is either male or female or the total bill is greater than or lesser than 30
```

```python
df[(df['day']=='Sun')|(df['day']=='Sat')|(df['day']=='Mon')] #Displays the entries where the day is either Saturday, Sunday or Monday
```

```python
options=['Sat','Sun','Thur'] #Creating a list of options on which we want our datas to be filtered
df['day'].isin(options) # Gets the boolean value of the dataframes in which the values in the option variables are present in day column of the dataframe
#df['day'].isin(options)==df['day'].isin(['Sat','Sun','Thur'])
```

```python
df[df['day'].isin(options)] #Displays the entires for which the day is in the list of options given
```

## MISSING DATA
* When reading in missing values,pandas will display them as NaN values.
Options for Missing Data:
    * Keep it
    * Remove it
    * Replace it

##### Keeping the missing data:
* Pros:
    * Easiest to do
    * Does not manipulate or change the true data
* Cons:
    * Many methods do not support NaN
    * Often there are reasonable guesses
 
##### Dropping or Removing the missing data:
* Pros:
    * Easy to do
    * Can be based on rules
* Cons:
    * Potential to lose a lot of data or useful information
    * Limits trained models for future data
    
    
* Drop a row - when a lot of informations are missing
* Often agood choice to calculate the percentage of what data is dropped
* Drop a column - if every row is missing that particular features

#### Filling in the missing data
* Pros:
    * Potential to save a lot of data for use in training a model
* Cons:
    * Hardest to do and somewhat arbitrary
    * Potential to lead to false conclusion


* Fill with same value - Good choice if Nan is a placeholder
* Fill with interpolated or estimated value - Much harder and requires reasonable assumptions

### MERGE
* When the Dataframes are not in the exact same order or format
* Types of Joins
    * Inner:
        * Selects all rows from both the dataframe as long as the conditions satisfies. The result-set will have rows from both the tables where the condition satisfies i.e value of the common field will be same.
            * pd.merge(dataframe1_name,dataframe2_name,how='inner',on='column_name')
    * Outer:
        * Creates the result-set by combining result of both left and right join. The result-set will contain all the rows from both the dataframe.The rows for which there is no matching the result-set will contain Null value.
            * pd.merge(dataframe1_name,dataframe2_name,how='outer',on='column_name')
    * Right :
        * Right join is similar to left join. This join returns all the rows of the dataframe on the right side of the join and matching rows for the dataframe on the left side of join. The rows for which there is no matching row on left side, the result-set will contain null. Right join is a.k.a Right Outer Join.
            * pd.merge(dataframe1_name,dataframe2_name,how='right',on='column_name')
    * Left:
        * This join returns all the rows of the dataframe on the left side of the join and matching rows for the dataframe on the right-side of the join. The rows for which there is no matching row on right side, the result set will contain null. Left join is a.k.a Left Outer Join.
            * pd.merge(dataframe1_name,dataframe2_name,how='left',on='column_name')

### PIVOT TABLE
* Pivot table allows us to reorganize data,refactopring cells based on columns and new index.
* A DataFrame with repeated values can be pivoted for a reorganization and clarity.

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Sales_Funnel_CRM.csv') #Reading in the CSV file and assigning it to a variable
df
```

```python
licenses=df[['Company','Product','Licenses']] # Creates a dataframe of the columns company, product and licenses
licenses
```

```python
pd.pivot(data=licenses,index='Company',columns='Product',values='Licenses') #gives how many number of licenses did we sell to a particular company
```

```python
pd.pivot_table(df,index='Company',aggfunc='sum') #Gives the aggregrate of the columns in the dataframe with Columns as the index. This code is equivalent to (df.groupby('Company').sum())
```

```python
pd.pivot_table(df,index='Company',aggfunc='sum',values=['Licenses','Sale Price']) #In the above code the Account Number was also summed up which wasn't logical. So we can pass the columns which we want to sum through the values arguement
```

```python
pd.pivot_table(df,index=['Account Manager','Contact'],values=['Sale Price'],aggfunc='sum') #Index=[First_Level_Index,Second_Level_Index],value is the parameter that we want as the output and the aggfunc is the action that has to be performed on the value based on the index categorization; this gives the total number of sale price made by the account manager per contact
```

```python
pd.pivot_table(df,index=['Account Manager','Contact'],values=['Sale Price'],columns=['Product'],aggfunc='sum') # This gives the sales price seperated out by Product
```

```python
pd.pivot_table(df,index=['Account Manager','Contact'],values=['Sale Price'],columns=['Product'],aggfunc=[np.sum,np.mean],fill_value=0) # To replace the NaN with 0; Note: we can send numpy function in aggregrate, this gives an advantage of having multiple function in one go, the result of that will be given as a new index 
```

```python
pd.pivot_table(df,index=['Account Manager','Contact'],values=['Sale Price'],columns=['Product'],aggfunc='sum',fill_value=0,margins=True,margins_name='Grand Total') # We can add a margin column at the end of the table which gives the sum along with the margin name 
```
